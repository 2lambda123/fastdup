## Downlaod & Extract Dataset
We will be using the [TV Human Interactions Video Dataset](https://www.robots.ox.ac.uk/~alonso/tv_human_interactions.html).

The dataset consists of 300 video clips collected from over 20 different TV shows and containing 4 interactions: hand shakes, high fives, hugs and kisses, as well as clips that don't contain any of the interactions.

```python
!wget http://www.robots.ox.ac.uk/~alonso/data/tv_human_interactions_videos.tar.gz
!tar xzf tv_human_interactions_videos.tar.gz
```

## Video to Images

fastdup works on images. We must first turn the videos into frames of images.

We can use a one-liner fastdup utility function to turn all the videos in a folder into frames:

```python
fastdup.extract_video_frames(input_dir="tv_human_interactions_videos", work_dir="frames")
```

## Run fastdup

```python
fd = fastdup.create(input_dir='frames', work_dir='face_detection_work_dir')
```

This is an advance functionality of fastdup. To use this function, sign up and get you license key at info@visual-layer.com

```python
fd.run(bounding_box='face', license='your_license_key', overwrite=True)
```

## Components Gallery

We can visualize the cluster of similar detections using the components gallery view.

```python
fd.vis.component_gallery(draw_bbox=True)
```

If you'd like to view just the cropped bounding box images, specify `draw_bbox=False`

```python
fd.vis.component_gallery(draw_bbox=False)
```

## Find Similar Detections Across Videos

```python
fd.vis.similarity_gallery(draw_bbox=False)
```
